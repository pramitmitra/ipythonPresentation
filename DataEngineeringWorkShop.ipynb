# What is Jupyter Notebook?

* The Jupyter Notebook is an interactive computing environment that enables users to author notebook documents that include: - Live code - Interactive widgets - Plots - Narrative text - Equations - Images - Video
* Enables code and text combination for user convenient. ### Components
* The notebook web application: An interactive web application for writing and running code interactively and authoring notebook documents.
* Kernels: Separate processes started by the notebook web application that runs usersâ€™ code in a given language and returns output back to the notebook web application.
* Notebook documents: Self-contained documents that contain a representation of all content visible in the notebook web application, including inputs and outputs of the computations, narrative text, equations, images, and rich media representations of objects. ### Notebook Notebook documents contain the only the input and output of the interactive session. When you run the notebook web application on your computer, notebook documents are just files on your local filesystem with a .ipynb extension. This format will serve us when submitting the HWs. These files are JSON files, which are rendered in the web browser

```python
import matplotlib.pyplot as plt
import math
import datetime as dt

test = """16:58:58 | 2.090 | 26.88  | 1.2945E-9  |   45.8\n
16:59:00 | 2.031 | 27.00  | 1.3526E-9  |  132.1\n
16:59:02 | 2.039 | 26.90  | 1.3843E-9  |  178.5\n
16:59:04 | 2.031 | 26.98  | 1.4628E-9  |  228.9\n
16:59:06 | 2.031 | 27.04  | 1.5263E-9  |  259.8\n
16:59:08 | 2.027 | 26.84  | 1.6010E-9  |  271.8\n"""

lines = [line for line in test.splitlines() if line != ""]

# Here is the real code
subset = []

for line in lines:
    parts = line.split('|')
    ts = dt.datetime.strptime(parts[0].strip(), "%H:%M:%S")
    num = math.log(float(parts[3].strip()))
    subset.append((ts, num))

# now there is a list of tuples with your datapoints, looking like
# [(datetime.datetime(1900, 1, 1, 16, 58, 58), 1.2945E-9), (datetime.datetime(1900, 1, 1, 16, 59), ...]
# I made this list intentionally so that you can see how one can gather everything in a tidy way from the
# raw string data.

# Now lets separate things for plotting
times = [elem[0] for elem in subset]
values = [elem[1] for elem in subset]

# now to plot, I'm going to use the matplotlib plot_date function.
plt.figure()
plt.plot_date(times, values)
# do some formatting on the date axis
plt.gcf().autofmt_xdate()
plt.show()
```

```python
import pandas

data = pandas.read_csv('data/population_UIAW.csv', index_col='CountryName')

# Extract year from last 4 characters of each column name
years = data.columns.str.strip('year_')
# Convert year values to integers, saving results back to dataframe
data.columns = years.astype(int)

data.loc[id].plot()
```

```python
data.T.plot()
plt.ylabel('Population capita')
```

```python
plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.ylabel('Population per country')
```

```python
years = data.columns
gdp_australia = data.loc['Australia']

plt.plot(years, gdp_australia, 'g--')
```

```python
# Select two countries' worth of data.
pop_australia = data.loc['Australia']
pop_usa = data.loc['USA']

# Plot with differently-colored markers.
plt.plot(years, pop_australia, 'b-', label='Australia')
plt.plot(years, pop_usa, 'g-', label='USA')

# Create legend.
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('Population')
```
